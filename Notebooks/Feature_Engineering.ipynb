```python
import pandas as pd
```

```python
df = pd.read_csv(r'cleaned_df.csv')
df
```

```python
df.drop(columns=['TransactionDate', 'TransactionTime'], axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['CustGender'] = encoder.fit_transform(df['CustGender'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.iloc[:, 2:] = scaler.fit_transform(df.iloc[:, 2:])
```

```python
location_encoder = LabelEncoder()
df['CustLocation'] = location_encoder.fit_transform(df['CustLocation'])
```

```python
df
```

```python
df.to_csv('feature_extracted.csv', index=False)
```
